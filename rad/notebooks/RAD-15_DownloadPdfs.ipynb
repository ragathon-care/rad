```python
import os
import requests
import os
import requests
from multiprocessing import Pool

# List of URLs pointing to PDFs
pdf_urls = list(pdf_url_df.child_url.unique())

# Directory where you want to save the downloaded PDFs
output_directory = 'downloaded_pdfs_multi'

# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Function to download a PDF and save it to the output directory
def download_pdf(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            # Extract the PDF file name from the URL
            pdf_name = os.path.basename(url)
            # Create the full path to save the PDF
            pdf_path = os.path.join(output_directory, pdf_name)
            # Save the PDF to the specified path
            with open(pdf_path, 'wb') as pdf_file:
                pdf_file.write(response.content)
            return url, pdf_name
        else:
            print(f"Failed to download PDF from {url}. Status code: {response.status_code}")
    except Exception as e:
        print(f"Error while downloading {url}: {str(e)}")

# Use multiprocessing to download PDFs concurrently
def download_pdfs_concurrently(pdf_urls):
    with Pool() as pool:
        results = pool.map(download_pdf, pdf_urls)
    return {url: pdf_name for url, pdf_name in results if pdf_name}

# Download PDFs and store their URL names
pdf_name_mapping = download_pdfs_concurrently(pdf_urls)

# Print the mapping of URLs to PDF names
for url, pdf_name in pdf_name_mapping.items():
    print(f"Downloaded {pdf_name} from {url}")

# Optionally, you can save the mapping to a text file
with open('pdf_name_mapping.txt', 'w') as mapping_file:
    for url, pdf_name in pdf_name_mapping.items():
        mapping_file.write(f"{url}: {pdf_name}\n")
```
