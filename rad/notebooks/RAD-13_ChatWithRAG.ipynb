```python
import pandas as pd
import os

# Assuming `df` is your DataFrame and it has columns 'url' and 'markdown_content'
# Make sure to replace 'your_dataframe.csv' with the actual path to your DataFrame file

# Specify the directory where you want to save the markdown files
output_directory = './final_artifacts/markdown_files/'

# Make sure the output directory exists, create if it doesn't
os.makedirs(output_directory, exist_ok=True)

for index, row in final_webpage_df.iterrows():
    # Extract URL and markdown content
    url = row['parent_url']
    markdown_content = row['markdown']
    
    # Define the filename (e.g., using the index or hashing the URL for uniqueness)
    filename = f"markdown_{index}.md"
    filepath = os.path.join(output_directory, filename)
    
    # Write the markdown content to a file
    with open(filepath, 'w', encoding='utf-8') as file:
        # Include a comment section with the original URL at the top
        file.write(f"<!-- Original URL: {url} -->\n\n")
        # Write the markdown content
        file.write(markdown_content)
    print(index)
print("Markdown files have been successfully created.")
```

```python
from llama_index import SimpleDirectoryReader
from llama_index.indices import VectaraIndex
```

```python
documents = SimpleDirectoryReader(os.path.abspath("./final_artifacts/markdown_files/")).load_data()
print(f"documents loaded into {len(documents)} document objects")
print(f"Document ID of first doc is {documents[0].doc_id}")
```

```python
index = VectaraIndex.from_documents(documents)
```

```python
query = "I have a child who recently had a stroke and is now recovery. They're going to high school soon. What should I know?"
```

```python
query_engine = index.as_query_engine(similarity_top_k=5)
response = query_engine.retrieve(query)
texts = [t.node.text for t in response]
print("\n--\n".join(texts))
```

```python
from llama_index import VectorStoreIndex, SimpleDirectoryReader, ServiceContext
from llama_index.llms import OpenAI

# Necessary to use the latest OpenAI models that support function calling API
service_context = ServiceContext.from_defaults(
    llm=OpenAI(model="gpt-3.5-turbo-0613")
)
data = SimpleDirectoryReader(input_dir="./final_artifacts/markdown_files/").load_data()
index = VectorStoreIndex.from_documents(data, service_context=service_context)
```

```python
chat_engine = index.as_chat_engine(chat_mode="openai", verbose=True)
```

```python
response = chat_engine.chat(
    f"Use the tool to answer: {query}?"
)
print(response)
```

```python
response = chat_engine.chat(
    f"Use the tool to answer: How can a 504 plan help me? What does it mean in California?"
)
print(response)
```
