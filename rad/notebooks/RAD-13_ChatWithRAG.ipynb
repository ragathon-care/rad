```python
from enum import Enum
from pydantic import BaseModel, Field
from typing import List, Optional
```

```python
from enum import Enum

class ResourceEnum(Enum): 
    form: "Form"
    informational: "Informational"
    other: "Other"

class AudienceEnum(Enum): 
    clientOrCaretaker: "Client or Caretaker"
    businessManager: "Business Manager"
    schoolCounselor: "School Counselor"
    socialWorker: "Social Worker"
    medicalProfessional: "Medical Professional"

class approxAgeGroup(Enum): 
    child: "Child"
    teen: "Teen"
    adult: "Adult"
    any: "Any"
    notApplicable: "Not Applicable"
```

```python
from typing import Literal
```

```python
class AudienceEntry(BaseModel): 
    entry: Literal['Client or Caretaker','Business Manager', 'School Counselor','Social Worker',' Medical Professional']
```

```python
class ResourceAnnotation(BaseModel): 
    category: Literal['Form','Informational','Other'] = Field(..., description="The category of the resource.")

    summary: str = Field(..., description="A one sentence summary of the resource.")

    primaryAudience: Literal['Person or Student with Disability', 'Business Manager', 'School Counselor','Social Worker',' Medical Professional', 'Other'] = Field(..., description="The primary audience for the resource.")

    secondaryAudience: Literal['Person or Student with Disability','Business Manager', 'School Counselor','Social Worker',' Medical Professional', 'N/A'] = Field(..., description="The secondary audience for the resource.")
```

```python
import pypdf
from pathlib import Path

def read_text_from_pdf(filepath):
    '''
    rough pass at reading text from pdf 
    '''
    file = Path(filepath)
    pdf = pypdf.PdfReader(file)

    # Get the number of pages in the PDF document
    num_pages = len(pdf.pages)

    # Iterate over every page
    all_text = []
    for page in range(num_pages):
        # Extract the text from the page
        page_text = pdf.pages[page].extract_text()
        all_text.append(page_text)
    return '\n'.join(all_text)
```

```python
import instructor
from openai import OpenAI

# Apply the patch to the OpenAI client
# enables response_model keyword
client = instructor.patch(OpenAI())
```

```python
def generate(data):
    query = f'''
Please explain, summarize, and annotate the pdf that I am providing to you. The pdf is demarcated by [STARTPDF] and [ENDPDF]. 

[STARTPDF]
{data}
[ENDPDF]

Now, please provide the requested information about the pdf below. 
'''
    return client.chat.completions.create(
        model="gpt-3.5-turbo",
        response_model=ResourceAnnotation,
        messages=[
            {
                "role": "system",
                "content": "You are an expert data annotator. Your job is to provide annotations on a pdf provided to you.",
            },
            {
                "role": "user",
                "content": query,
            },
        ],
    )  # type: ignore
```

```python
def annotate_pdf(sample_path):
    try: 
        sample_text = read_text_from_pdf(sample_path)
        out = generate(sample_text)
        out.__dict__.pop('_raw_response')
        out.__dict__['source'] = sample_path
        return out.__dict__
    except: 
        return {'source':sample_path}
```
